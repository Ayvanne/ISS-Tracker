```python
import requests
import matplotlib.pyplot as plt
```

```python
url = "http://api.open-notify.org/iss-now.json"
```

```python
latitudes = []
longitudes = []
timestamps = []
```

```python
import time

data_points = []

for _ in range(100):
    response = requests.get(url)
    data = response.json()
    latitude = float(data["iss_position"]["latitude"])
    longitude = float(data["iss_position"]["longitude"])
    timestamp = int(data["timestamp"])
    latitudes.append(latitude)
    longitudes.append(longitude)
    timestamps.append(timestamp)
    data_points.append((latitude, longitude, timestamp))
    print(f"Latitude: {latitude}, Longitude: {longitude}, Timestamp: {timestamp}")
    time.sleep(10)
```

```python
plt.scatter(longitudes, latitudes, c=timestamps, cmap="coolwarm", s=10)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("International Space Station Path")
plt.colorbar(label="Timestamp")
plt.show()
```
